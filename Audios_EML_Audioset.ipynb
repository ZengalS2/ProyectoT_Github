Descarga masiva de audios de Youtube

```python
destination_bt_cry = "K:\Proyecto T\Audioset\Balanced Training\Crying audios"
destination_ev_cry = "K:\Proyecto T\Audioset\Evaluation\Crying audios"
destination_bt_notcry = "K:\Proyecto T\Audioset\Balanced Training\\NotCrying audios"
destination_ev_notcry = "K:\Proyecto T\Audioset\Evaluation\\NotCrying audios"

dest_list = [destination_bt_cry, destination_bt_notcry, destination_ev_cry, destination_ev_notcry]

#Se disminuye los dataframes mas grades (aquellos de training y evaluacion no pertenecientes 
#a gritos) a las primeras 1500 filas o rows
df_notcry_bt_cut = df_notcry_bt.head(1500)
df_notcry_ev_cut = df_notcry_ev.head(1500)

df_list = [df_cry_bt, df_notcry_bt_cut, df_cry_ev, df_notcry_ev_cut]

```

```python
for index_dd in range(len(dest_list)):
    print(dest_list[index_dd])
```

```python
#df.loc[df["name"]=="Crying, sobbing"]
df_cry_bt.loc[df_cry_bt["YTID"] == "zQpPfONDPXs"]
```

```python
for index_dest in range(len(dest_list)):

    error_list = []

    if not os.path.exists(dest_list[index_dest]):
        os.makedirs(dest_list[index_dest])

    df_temp = df_list[index_dest]
    for i, row in df_temp.iterrows():
        try:
            if not os.path.exists(dest_list[index_dest]+"\\"+row[0]+".wav"):
                yt_audio.DescargaYT_formato_corte(row[0], row[1] * 1000, row[2] * 1000, dest_list[index_dest])
        except:    
            error_list.append(row[0])
            continue
    
    errortxt = open(dest_list[index_dest]+"\\errors.txt", "w")
    errortxt.truncate(0)

    print("Errores en los siguientes YTIDs: ", error_list)
    for string in error_list:
        errortxt.write(string + "\n")
        
    errortxt.close()
```
